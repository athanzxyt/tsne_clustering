```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier as DT
from sklearn.linear_model import LogisticRegression as LR
from sklearn.svm import SVC as SVM
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import train_test_split

from sklearn.manifold import TSNE
from tsne import tsne

SEED = 1234
np.random.seed=(SEED)
```

## MNIST

```python
# MNIST Data
mnist_data = np.loadtxt("data/mnist/mnist2500_X.txt")
mnist_labels = np.loadtxt("data/mnist/mnist2500_labels.txt")
mnist_data_tsne = TSNE(n_components=3, 
                       learning_rate='auto', 
                       init='pca', 
                       perplexity=30,
                       random_state=SEED).fit_transform(mnist_data)
```

```python
for label in np.unique(mnist_labels):
    plt.scatter(mnist_data_tsne[:,0][mnist_labels == label],
                mnist_data_tsne[:,1][mnist_labels == label],
                s=5,
                label=int(label))
plt.legend(bbox_to_anchor=(1.2, 1.0), loc='upper right')
plt.xlabel("Dim 1")
plt.ylabel("Dim 2")
# plt.title("Perplexity = ")
plt.show()
```

### No t-SNE Classification

```python
X_train, X_test, y_train, y_test = train_test_split(mnist_data, mnist_labels, test_size=0.2, random_state=SEED)

classifiers = dict()
classifiers["DT"]   = DT(random_state=SEED)
classifiers["LR"]   = LR(random_state=SEED, max_iter=5000)
classifiers["SVM"]  = SVM(random_state=SEED)
classifiers["KNN"]  = KNN()

for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    print("MNIST with %s Acc: %s" % (name, round(accuracy_score(y_test, preds),3)))
    print("MNIST with %s F1: %s" % (name, round(f1_score(y_test, preds,average='weighted'),3)))
```

### With t-SNE Classification

```python
X_train, X_test, y_train, y_test = train_test_split(mnist_data_tsne, mnist_labels, test_size=0.2, random_state=SEED)

classifiers = dict()
classifiers["DT"]   = DT(random_state=SEED)
classifiers["LR"]   = LR(random_state=SEED, max_iter=5000)
classifiers["SVM"]  = SVM(random_state=SEED)
classifiers["KNN"]  = KNN()

for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    print("MNIST with tsne+%s Acc: %s" % (name, round(accuracy_score(y_test, preds),3)))
    print("MNIST with tsne+%s F1: %s" % (name, round(f1_score(y_test, preds,average='weighted'),3)))
```

## CIFAR10

```python
import pickle
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='latin1')
    return dict
```

```python
# Cifar10 Data
data_batch_1 = unpickle('data/cifar10/data_batch_1')
cifar10_data = np.array(data_batch_1['data'])
cifar10_labels = np.array(data_batch_1['labels'])
cifar10_data_tsne = TSNE(n_components=2, 
                       learning_rate='auto', 
                       init='pca', 
                       perplexity=100,
                       random_state=SEED).fit_transform(cifar10_data)
```

```python
for label in np.unique(cifar10_labels):
    plt.scatter(cifar10_data_tsne[:,0][cifar10_labels == label],
                cifar10_data_tsne[:,1][cifar10_labels == label],
                s=5,
                label=int(label))
plt.legend(bbox_to_anchor=(1.2, 1.0), loc='upper right')
plt.xlabel("Dim 1")
plt.ylabel("Dim 2")
# plt.title("Perplexity = ")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cifar10_data, cifar10_labels, test_size=0.2, random_state=SEED)

classifiers = dict()
classifiers["DT"]   = DT(random_state=SEED)
classifiers["LR"]   = LR(random_state=SEED, max_iter=5000)
classifiers["SVM"]  = SVM(random_state=SEED)
classifiers["KNN"]  = KNN()

for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    print("CIFAR10 with %s Acc: %s" % (name, round(accuracy_score(y_test, preds),3)))
    print("CIFAR10 with %s F1: %s" % (name, round(f1_score(y_test, preds,average='weighted'),3)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(cifar10_data_tsne, cifar10_labels, test_size=0.2, random_state=SEED)

classifiers = dict()
classifiers["DT"]   = DT(random_state=SEED)
classifiers["LR"]   = LR(random_state=SEED, max_iter=5000)
classifiers["SVM"]  = SVM(random_state=SEED)
classifiers["KNN"]  = KNN()

for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    print("CIFAR10 with tsne+%s Acc: %s" % (name, round(accuracy_score(y_test, preds),3)))
    print("CIFAR10 with tsne+%s F1: %s" % (name, round(f1_score(y_test, preds,average='weighted'),3)))
```
